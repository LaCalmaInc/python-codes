```python
data.head()
```

```python
top10clubAssistants = data.sort_values(by='Assists', ascending=False).head(10)
top10clubAssistants
```

```python
top10clubBallers = data.sort_values(by='Goals',ascending=False).head(10)
top10clubBallers
```

```python
filtered_data  = data.sort_values(by='Goal Participation Per Match',ascending=False)
sorted_data = filtered_data[filtered_data['Matches Played']>30]

unique_players = sorted_data.drop_duplicates(subset='Name')

# Seleccionar el top 10 jugadores
top10_gppm = unique_players.head(10)
top10_gppm
```

```python
nuevo_df = data[data['Name'].isin(top10_gppm['Name'])]
nuevo_df
```

```python
dataset_grafico = nuevo_df.pivot_table(index='Year',columns='Name',values='Goal Participation Per Match')
dataset_grafico
```

```python
dataset_grafico.to_csv('dataset_grafico.csv')
```
